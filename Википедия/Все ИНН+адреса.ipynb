```python
burn_copy = burn[['Адрес', 'ИНН участника']]
```

```python
burn_copy.rename(
    columns={'Адрес':'адрес', 'ИНН участника':'инн'},
    inplace=True
)
```

```python
full_addr = pd.concat([dsmh_copy, burn_copy])
```

```python
full_addr.drop_duplicates(inplace=True)
```

```python
full_addr = full_addr.loc[full_addr['адрес'].isna() == False]
```

```python
full_addr
```

```python
kb = pd.read_excel(r"D:\Analysis Burn\Википедия\Справочники\Адреса КБ\Адреса КБ.xlsx")
```

```python
full_addr
```

```python
add_and_inn = full_addr.groupby(
    ['адрес']
)['инн'].count().sort_values(ascending=False).reset_index()
```

```python
add_and_inn = add_and_inn.loc[add_and_inn['инн'] == 1]
```

```python
add_and_inn = add_and_inn.merge(
    full_addr,
    how='left',
    on = 'адрес'
)
```

```python
add_and_inn = add_and_inn.loc[add_and_inn['инн_y'].isna() == False]
```

```python
kb.loc[kb['инн_y'].isna()]
```

```python
addresses = pd.read_excel(r"D:\Analysis Burn\Википедия\Справочники\addresses.xlsx", sheet_name='Основная')
```

```python
addresses = addresses[['Исходный адрес', 'Адрес']]
```

```python
kb = kb.merge(add_and_inn[['адрес', 'инн_y']], 
    how='left',
    left_on=['Адрес_корректный'],
    right_on=['адрес']
)
```

```python
kb = kb.merge(
    addresses,
    how='left',
    left_on='Адрес_корректный',
    right_on='Адрес'
)
```

```python
kb_copy = kb.merge(add_and_inn[['адрес', 'инн_y']], 
    how='left',
    left_on=['Исходный адрес'],
    right_on=['адрес']
)
```
