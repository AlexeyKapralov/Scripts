```python
# добавить год  и месяци
df['Дата отгрузки'] = pd.to_datetime(df['Дата отгрузки'], dayfirst=True)

df['year'] = df['Дата отгрузки'].dt.year
df['month'] = df['Дата отгрузки'].dt.month
```

```python
# оставить только москву
df = df.loc[
    (df['Торговый представитель'] != 'REG MAN 001 (Такаев: ЮФО, СНГ)' ) & 
    (df['Торговый представитель'] != 'REG MAN 005 (Цыганков: ДВФО, ПФО)') &
    (df['Торговый представитель'] != 'REG MAN 004 (Кособокова: ЦФО, СЗФО)') &
    (df['Торговый представитель'] != 'REG MAN 006 (Олейник: УрФО, СФО)') &
    (df['Торговый представитель'] != 'Дудырев Владимир') & 
    (df['Торговый представитель'] != 'Гильманов Тимур') &
    (df['Торговый представитель'] != 'Гусенцов Александр') &
    (df['Торговый представитель'] != 'Telesales 002 (Юшкин)') &
    (df['Торговый представитель'] != 'Telesales 001 (Маслов)') &
    (df['Торговый представитель'] != 'Садыков Александр') &
    (df['Торговый представитель'] != 'INC MAN 002 (Маслов)') &
    (df['Торговый представитель'] != 'INC MAN 004 (Юшкин)') &
    (df['Торговый представитель'] != 'REG MAN 007 (Меркурьев)') &
    (df['Торговый представитель'] != 'Telesales 003 (Творогов)') &
    (df['Торговый представитель'] != 'Фаткин Ким') &
    (df['Торговый представитель'] != 'Лубинец Дмитрий' ) &
    (df['Торговый представитель'] != 'Абильфазов Азим' ) &
    (df['Торговый представитель'] != 'Асланян Ашхен' ) &
    (df['Торговый представитель'] != 'Баранов Алексей' ) & 
    (df['Торговый представитель'] != 'Сафронов Николай') 
] 
```

```python
df = df.loc[
    (df['Сегмент'] == 'Кальянная') |
    (df['Сегмент'] == 'HoReCa') |
    (df['Сегмент'] == 'Бар/Ночной клуб') |
    (df['Сегмент'] == 'Кафе/Ресторан') |
    (df['Сегмент'] == 'кальянная') |
    (df['Сегмент'] == 'horeca')
]
```

```python
pt = df.pivot_table(
    index=['Бренд','Вкус','Код клиента'],
    columns=['year', 'month'],
    values='ВЕС',
    aggfunc=lambda x: 1 if sum(x) > 0 else 0
).reset_index()
```

```python
# преобразование multi-index в обычный
pt.columns = ['_'.join(map(str, col)).rstrip('_') for col in pt.columns.values]
```
