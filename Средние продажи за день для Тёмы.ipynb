```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Analysis Burn\Дистрибьютеры отчетность\Source\Отчеты\Москва OSHISHA\Sell_out_Москва_OSHISHA.csv")
```

```python
one = df.loc[df['Сегмент'] == 'табаконист']
```

```python
# преобразование типа в дату
def convert_date(date_series):
    if date_series[0:4].isdigit():
        # Попытка преобразования с yearfirst=True
        converted_dates = pd.to_datetime(date_series, yearfirst=True)
    else:
        # Если возникла ошибка, пробуем преобразовать с dayfirst=True
        converted_dates = pd.to_datetime(date_series, dayfirst=True)
    return converted_dates


if one['Дата отгрузки'].dtypes != '<M8[ns]':
    one['Дата отгрузки'] = one['Дата отгрузки'].apply(convert_date)
```

```python
one['month'] = one['Дата отгрузки'].dt.month
one['year'] = one['Дата отгрузки'].dt.year
```

```python
one = one.loc[(one['Бренд'] == 'BlackBurn') | (one['Бренд'] == 'Blackburn')]
```

```python
one['day'] = one['Дата отгрузки'].dt.day
```

```python
one['Код клиента'] = one['Код клиента'].str.lower()
```

```python
one = one.groupby(
    by=['Код клиента', 'month',	'year',	'day']
)['ВЕС'].sum().reset_index()
```

```python
one_two = one.copy(deep=True)
one_two['кол_во'] = 1
```

```python
three = one_two.pivot_table(
    values= 'кол-во',
    index=['Код клиента'],
    columns=['year', 'month'],
    aggfunc="sum"
).reset_index()
```

```python
three['Кол_во'] = three.iloc[:,1:].mean(axis=1)
```

```python
three.drop(columns=three.iloc[:,1:-1], inplace=True)
```

```python
two = one.pivot_table(
    values= 'ВЕС',
    index=['Код клиента'],
    columns=['year', 'month', 'day']
).reset_index()
```

```python
two['Средняя в день'] = two.iloc[:,1:].mean(axis=1)
two['Сумма'] = two.iloc[:,1:].sum(axis=1)
```

```python
two.drop(columns=two.iloc[:,1:-2], inplace=True)
```
