```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Analysis Burn\Дистрибьютеры отчетность\Source\Отчеты\Москва OSHISHA\Sell_out_Москва_OSHISHA.csv")
```

```python
if df['Дата отгрузки'].str[:4].str.match(r"\d{4}.*",).all():
    df['Дата отгрузки'] = pd.to_datetime(df['Дата отгрузки'], yearfirst=True)
```

```python
df['month'] = df['Дата отгрузки'].dt.month
df['year'] = df['Дата отгрузки'].dt.year
```

### Создание сводной таблицы продаж в кг

```python
general_weight = df.pivot_table(
    values='ВЕС',
    index='Сегмент',
    columns=['year','month'],
    aggfunc = 'sum'
).reset_index()
```

```python
general_weight.rename(columns={1:"янв"}, inplace=True)
general_weight.rename(columns={2:"фев"}, inplace=True)
general_weight.rename(columns={3:"мар"}, inplace=True)
general_weight.rename(columns={4:"апр"}, inplace=True)
general_weight.rename(columns={5:"май"}, inplace=True)
general_weight.rename(columns={6:"июн"}, inplace=True)
general_weight.rename(columns={7:"июл"}, inplace=True)
general_weight.rename(columns={8:"авг"}, inplace=True)
general_weight.rename(columns={9:"сен"}, inplace=True)
general_weight.rename(columns={10:"окт"}, inplace=True)
general_weight.rename(columns={11:"ноя"}, inplace=True)
general_weight.rename(columns={12:"дек"}, inplace=True)
```

```python
general_weight.columns = [f'{i}{"-"}{j}' for i, j in general_weight.columns]
```

### Создание сводной таблицы продаж в долях каждого сегмента в %

```python
general_proportion = df.pivot_table(
    values='ВЕС',
    index='Сегмент',
    columns=['year','month'],
    aggfunc= 'sum'
).reset_index()
```

```python
general_proportion.rename(columns={1:"янв"}, inplace=True)
general_proportion.rename(columns={2:"фев"}, inplace=True)
general_proportion.rename(columns={3:"мар"}, inplace=True)
general_proportion.rename(columns={4:"апр"}, inplace=True)
general_proportion.rename(columns={5:"май"}, inplace=True)
general_proportion.rename(columns={6:"июн"}, inplace=True)
general_proportion.rename(columns={7:"июл"}, inplace=True)
general_proportion.rename(columns={8:"авг"}, inplace=True)
general_proportion.rename(columns={9:"сен"}, inplace=True)
general_proportion.rename(columns={10:"окт"}, inplace=True)
general_proportion.rename(columns={11:"ноя"}, inplace=True)
general_proportion.rename(columns={12:"дек"}, inplace=True)
```

```python
general_proportion.columns = [f'{i}{"-"}{j}' for i, j in general_proportion.columns]
```

```python
for a in general_proportion.iloc[:,1:].columns:
    general_proportion[a] = general_proportion[a] / general_proportion[a].sum()
```

### Создание сводной таблицы Средний чек

```python
df.head(2)
```

```python
sr_check_temp = df.groupby(by=['Сегмент','year', 'month']).agg({'ВЕС': 'sum', 'Код клиента':'nunique'}).reset_index()
```

```python
sr_check_temp['Средний чек'] = sr_check_temp['ВЕС'] / sr_check_temp['Код клиента']
sr_check_temp['Средний чек'] = sr_check_temp['Средний чек'].round(2)
```

```python
sr_check_temp.head(1)
```

```python
sr_check = sr_check_temp.pivot_table(
    values='Средний чек',
    index='Сегмент',
    columns=['year','month'],
    aggfunc='sum'
).reset_index()
```

```python
sr_check.rename(columns={1:"янв"}, inplace=True)
sr_check.rename(columns={2:"фев"}, inplace=True)
sr_check.rename(columns={3:"мар"}, inplace=True)
sr_check.rename(columns={4:"апр"}, inplace=True)
sr_check.rename(columns={5:"май"}, inplace=True)
sr_check.rename(columns={6:"июн"}, inplace=True)
sr_check.rename(columns={7:"июл"}, inplace=True)
sr_check.rename(columns={8:"авг"}, inplace=True)
sr_check.rename(columns={9:"сен"}, inplace=True)
sr_check.rename(columns={10:"окт"}, inplace=True)
sr_check.rename(columns={11:"ноя"}, inplace=True)
sr_check.rename(columns={12:"дек"}, inplace=True)
```

```python
sr_check.columns = [f'{i}{"-"}{j}' for i, j in sr_check.columns]
```

### Потерянные клиенты

```python
df.head(1)
```

```python
lost_clients = df.pivot_table(
    values='ВЕС',
    columns = ['year', 'month'],
    index = ['Сегмент','Код клиента'],
    aggfunc=lambda x: 1 if sum(x) > 0 else 0,
    fill_value=0
).reset_index()
```

```python
lost_clients.rename(columns={1:"янв"}, inplace=True)
lost_clients.rename(columns={2:"фев"}, inplace=True)
lost_clients.rename(columns={3:"мар"}, inplace=True)
lost_clients.rename(columns={4:"апр"}, inplace=True)
lost_clients.rename(columns={5:"май"}, inplace=True)
lost_clients.rename(columns={6:"июн"}, inplace=True)
lost_clients.rename(columns={7:"июл"}, inplace=True)
lost_clients.rename(columns={8:"авг"}, inplace=True)
lost_clients.rename(columns={9:"сен"}, inplace=True)
lost_clients.rename(columns={10:"окт"}, inplace=True)
lost_clients.rename(columns={11:"ноя"}, inplace=True)
lost_clients.rename(columns={12:"дек"}, inplace=True)
```

```python
lost_clients.columns = [f'{i}{"-"}{j}' for i, j in lost_clients.columns]
```

```python
b = [i for i in range(2, lost_clients.shape[1])]
```

```python
b
```

```python
lost_clients
```

```python
for i in b:

        if i == 2 or i == 3 or i == 4: 
            lost_clients[lost_clients.columns[i] + '-2'] = 0
            continue

        lost_clients[lost_clients.columns[i] + '-2'] = lost_clients.iloc[:,i-3:i+1].sum(axis=1)

lost_clients.drop(lost_clients.columns[b[0]:b[-1]+1], axis= 1, inplace=True)

lost_clients.columns = lost_clients.columns.str.replace('-2', '',regex=True)
            
```

```python
lost_clients.head(2)
```

```python
for i in lost_clients.iloc[:,5:].columns:

    lost_clients[i] = np.where(lost_clients[i] > 0, 0, 1)
```

```python
lost_clients.to_excel(r"C:\Users\User\Desktop\lost_clients_new.xlsx")
```

```python
lost_clients_temp = pd.melt(
    lost_clients,
    id_vars=['Сегмент-', 'Код клиента-'],
    value_vars=lost_clients.iloc[:,2:].columns.values
)
```

```python
lost_clients_temp = lost_clients_temp.groupby(by=['Сегмент-', 'variable'])['value'].sum().reset_index()
```

```python
lost_clients_temp = lost_clients_temp.pivot_table(
    values = 'value',
    index = 'Сегмент-',
    columns = 'variable'
).reset_index()
```

### Сохранение в таблицу

```python
with pd.ExcelWriter(r"C:\Users\User\Desktop\Business_review.xlsx") as writer: 
    general_weight.to_excel(writer, sheet_name='Общие продажи (кг)', index=False) 
    general_proportion.to_excel(writer, sheet_name='Общие продажи (%)', index=False)
    akb.to_excel(writer, sheet_name='АКБ', index=False)
    sr_check.to_excel(writer, sheet_name='Средний чек', index=False)
    lost_clients_temp.to_excel(writer, sheet_name='Потерянные клиенты', index=False)
```
