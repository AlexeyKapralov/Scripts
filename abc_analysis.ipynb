```python
import pandas as pd
```

```python
banger = pd.read_csv(r"D:\Analysis Burn\Дистрибьютеры отчетность\Source\Отчеты\Москва OSHISHA\Sell_out_Москва_OSHISHA.csv")
```

```python
overdose = pd.read_csv(r"D:\Analysis Burn\Дистрибьютеры отчетность\Source\Отчеты\Москва OSHISHA\Sell_out_Москва_OSHISHA.csv")
```

```python
banger['Дата отгрузки'] = pd.to_datetime(banger['Дата отгрузки'], yearfirst=True)
```

```python
banger['month'] = banger['Дата отгрузки'].dt.month
banger['year'] = banger['Дата отгрузки'].dt.year
```

```python
df1_banger = banger.loc[(banger.month == 6) & (banger['Бренд'] == 'Banger') & (banger['Группа'] == 'Banger 25 гр')]
```

```python
df1_banger.head(2)
```

```python
df1_banger = df1_banger.groupby('Вкус').agg({'ВЕС':'sum'}).reset_index().sort_values('ВЕС',ascending=False)
```

```python
df1_banger_sum = df1_banger['ВЕС'].sum()
```

```python
df1_banger['Доля %'] = (((df1_banger['ВЕС'] / df1_banger_sum)).round(3) * 100)
```

```python
df1_banger['Доля %'].sum()
```

```python
df1_banger
```

```python
overdose['Дата отгрузки'] = pd.to_datetime(overdose['Дата отгрузки'], yearfirst=True)
```

```python
overdose['month'] = overdose['Дата отгрузки'].dt.month
overdose['year'] = overdose['Дата отгрузки'].dt.year
```

```python
df1_overdose = overdose.loc[(overdose.month == 6) & (overdose['Бренд'] == 'Overdose') & (overdose['Группа'] == 'Overdose 25 гр')]
```
