```python
daily_data2 = daily_data.pivot_table(
    index=['Вкус', 'Код клиента'],
    columns='Дата отгрузки',
    values='Код клиента',
    aggfunc=lambda x: 1 if sum(x) > 0 else 0
).reset_index()
```

```python
daily_data2.columns = daily_data2.columns.astype(str)
```

```python
col_size = daily_data2.columns.size
```

```python
row_size = len(daily_data2)-1
```

```python
i = 2
iterator = 3

while iterator <= col_size:
    name = daily_data2.columns[i]
    daily_data2[name.replace('00:00:00',"")] = daily_data2.iloc[:, 2:iterator].sum(axis=1)
    i += 1
    iterator += 1
```

```python
daily_data2 = daily_data2.drop(daily_data2.columns[2:col_size], axis=1)
```

```python
import numpy as np
# Определяем функцию для фильтрации значений
def filter_values(value):
    if value == 1:
        return value
    else:
        return np.nan

daily_data2.iloc[:, 2:] = daily_data2.iloc[:, 1:].applymap(filter_values)
```
