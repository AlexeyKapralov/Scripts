```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Datasets\diamonds.csv")
```

```python
df
```

```python
df_prepared = df.groupby(
    by=['color', 'cut']
).size().unstack()
```

```python
df_prepared
```

```python
import matplotlib.pyplot as plt

# Построение гистограммы
fig, ax =plt.subplots(figsize=(14, 7))
df_prepared.plot(kind='bar', ax=ax)

ax.set_title('Количество бриллиантов по Цвету и Огранке')
ax.set_xlabel('Color')
ax.set_ylabel('Count')
ax.legend(title='Cut')

# Показать график
plt.tight_layout()
plt.show()
```
