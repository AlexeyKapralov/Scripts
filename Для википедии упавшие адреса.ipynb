```python
import pandas as pd
from datetime import date
```

```python
wiki = pd.read_excel(r"D:\Analysis Burn\Википедия\Wikipedia.xlsx")
```

```python
current_year = date.today().year
current_month = date.today().month - 1
current_month_2 = date.today().month - 2
```

```python
wiki_filter = wiki.loc[ (wiki['Год'] == current_year) & ( (wiki['Месяц'] == current_month_2) | (wiki['Месяц'] == current_month) ) ]
```

```python
wiki_filter = wiki_filter.pivot_table(
    index=['Субъект РФ', 'Адрес', 'Бренд'],
    columns=['Месяц'],
    values=['Вес'],
    aggfunc='sum'
).reset_index()
```

```python
# Объединяем значения кортежей, пропуская пустые
joined_values = ['_'.join([str(x) for x in idx if x]) for idx in wiki_filter]
```

```python
wiki_filter.columns = joined_values
```

```python
name_pre_last_column = wiki_filter.iloc[:,-2:-1].isna().columns[0]
name_last_column = wiki_filter.iloc[:,-1:].isna().columns[0]
```

```python
wiki_filter.loc[ wiki_filter[name_pre_last_column].isna(), name_pre_last_column] = 0

wiki_filter.loc[ wiki_filter[name_last_column].isna(), name_last_column] = 0
```

```python
wiki_filter
```

```python
wiki_filter['Разница с предыдующим месяцем'] = wiki_filter[name_last_column] - wiki_filter[name_pre_last_column]
```

```python
wiki_filter.pivot_table(
    index=["Субъект РФ", "Адрес"],
    columns=["Бренд"],
    values=['Разница с предыдующим месяцем'],
    aggfunc='sum'
).reset_index()
```
